```python
# Aggre Transaction Details

Path1="D:/Phone pay/pulse/data/aggregated/transaction/country/india/state/"
agg_trans_list=os.listdir(Path1)

column1={'States':[], 'Years':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for state in agg_trans_list:
    state_1=Path1+state+"/"
    Agg_yr=os.listdir(state_1)
    for years in Agg_yr:
        years_1=state_1+years+"/"
        Agg_yr_list=os.listdir(years_1)
        for file in Agg_yr_list:
            file_1=years_1+file
            Data=open(file_1,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              column1['Transaction_type'].append(Name)
              column1['Transaction_count'].append(count)
              column1['Transaction_amount'].append(amount)
              column1['States'].append(state)
              column1['Years'].append(years)
              column1['Quarter'].append(int(file.strip('.json')))
#Succesfully created a dataframe
Agg_Trans=pd.DataFrame(column1)

Agg_Trans["States"] = Agg_Trans["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Agg_Trans["States"] = Agg_Trans["States"].str.replace("-"," ")
Agg_Trans["States"] = Agg_Trans["States"].str.title()
Agg_Trans['States'] = Agg_Trans['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

```

```python
# Aggre user details

Path2="D:/Phone pay/pulse/data/aggregated/user/country/india/state/"
Agg_user_list=os.listdir(Path2)

column2={'States':[], 'Years':[],'Quarter':[],'Brands':[], 'Transaction_count':[], 'Percentage':[]}

for state in Agg_user_list:
    state_2=Path2+state+"/"
    Agg_yr=os.listdir(state_2)
    for years in Agg_yr:
        years_2=state_2+years+"/"
        Agg_yr_list=os.listdir(years_2)
        for file in Agg_yr_list:
            file_2=years_2+file
            Data2=open(file_2,"r")
            E=json.load(Data2)

            try:
                for z in E['data']['usersByDevice']:
                    brands=z['brand']
                    count=z['count']
                    percentage=z['percentage']
                    column2['Brands'].append(brands)
                    column2['Transaction_count'].append(count)
                    column2['Percentage'].append(percentage)
                    column2['States'].append(state)
                    column2['Years'].append(years)
                    column2['Quarter'].append(int(file.strip('.json')))

            except:
                pass

Agg_user=pd.DataFrame(column2)

Agg_user["States"] = Agg_user["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Agg_user["States"] = Agg_user["States"].str.replace("-"," ")
Agg_user["States"] = Agg_user["States"].str.title()
Agg_user['States'] = Agg_user['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")


    
```
